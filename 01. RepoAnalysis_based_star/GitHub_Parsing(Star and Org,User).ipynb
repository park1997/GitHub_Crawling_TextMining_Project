```python
import requests
from bs4 import BeautifulSoup
import lxml
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("API_info_gathering.xlsx")
df.columns = ['api']
df
```

```python
df['api']
```

```python
import json
dic_str = ""
for i in df['api']:
    dic_str += i
    print(i)
df_json = json.loads(dic_str)
df_json
```

```python
name = []
star = []
owner_type = []
info = {}
for i in df_json['items']:
#     print(i['name'], i['stargazers_count'], i['owner']['type'])
    name.append(i['name'])
    star.append(i['stargazers_count'])
    owner_type.append(i['owner']['type'])
    info['name'] = name
    info['star'] = star
    info['owner_type'] = owner_type
```

```python
df_parse = pd.DataFrame(info)
df_parse
```

```python
df_parse.to_excel('github_stars_info.xlsx')
```
