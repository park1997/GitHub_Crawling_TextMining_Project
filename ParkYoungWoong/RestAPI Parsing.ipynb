```python
import json
from urllib.request import urlopen
```

```python
response = urlopen('https://api.github.com/search/repositories?q={deep%20learning}{&page,per_page,sort,order}').read().decode('utf-8')
```

```python
responseJson = json.loads(response)
responseJson
```

```python
items = responseJson.get('items')

for lst in items:
    name = lst.get('name')
    html = lst.get('html_url')
    desc = lst.get('description')
    typ = lst.get('owner').get('type')
    
    print(name,'/', html,'/', desc,'/', typ, '\n')
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

driver = webdriver.Chrome('C:/chromedriver_win32/chromedriver', options = options)
driver.implicitly_wait(3)

driver.get('https://github.com')
driver.implicitly_wait(3)

driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/div[2]/div[1]/div/div/form/label/input[1]').clear()
driver.implicitly_wait(3)
```

```python
def hasclass(name):
    try: 
        driver.find_element_by_class_name(name)
        return True
    except:
        return False
```
